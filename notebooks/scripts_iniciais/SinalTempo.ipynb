```python
import matplotlib.pyplot as plt
import librosa
import librosa.display
import numpy as np
import os

def plot_sinal_tempo(file_path, sr=None, title=None):
    """
    Plota o sinal de áudio no domínio do tempo.
    
    Parâmetros:
    - file_path: caminho do arquivo .wav
    - sr: sample rate desejado (None para manter original)
    - title: título do gráfico (opcional)
    """
    y, sr = librosa.load(file_path, sr=sr)
    duration = librosa.get_duration(y=y, sr=sr)
    time = np.linspace(0, duration, len(y))
    
    plt.figure(figsize=(12, 4))
    plt.plot(time, y, color='dodgerblue')
    plt.xlabel("Tempo (s)")
    plt.ylabel("Amplitude")
    plt.title(title or f"Sinal no tempo – {os.path.basename(file_path)}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
file_exemplo = "./2.wav"
plot_sinal_tempo(file_exemplo)
```

```python
file_exemplo = "./segment_3.wav"
plot_sinal_tempo(file_exemplo)
```
